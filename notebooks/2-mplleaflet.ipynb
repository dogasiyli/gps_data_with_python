```python
import gpxpy
import mplleaflet
import pandas as pd

import matplotlib.pyplot as plt
plt.rcParams['axes.xmargin'] = 0.1
plt.rcParams['axes.ymargin'] = 0.1
%matplotlib inline

import seaborn as sns
sns.set_style("whitegrid")
sns.set_context("talk")
```

```python
from helpers import *
gpx_files = list_gpx_files()
gpx_file = load_gpx_file(gpx_files, 2)
```

File consists of many tracks und track consists of segments which consists of points.
Segments are continuously connected points. Points in different segments are not connected to reflect loss of gps signal, device turned on/off etc.

```python
segment = gpx_file.tracks[0].segments[0]
coords = pd.DataFrame([{'lat': p.latitude, 'lon': p.longitude} for p in segment.points])
plot_coords = coords.iloc[::150]
```

## How about plotting the track?

```python
fig = plt.figure()
plt.plot(plot_coords['lon'].values, plot_coords['lat'].values)
plt.plot(plot_coords['lon'].values, plot_coords['lat'].values, 'ro')
```

## And now on OpenStreeMap!

```python
mplleaflet.display(fig=fig)
```
